###### Finding Outliers

```python
##finding outliers in popularity column
plt.figure(figsize=(15,6))
plt.title('Outliers of Popularity')
sns.boxplot(df['popularity'])
plt.xticks(rotation = 90)
plt.show()
```

```python
popularity_d=df['popularity']
Q3=popularity_d.quantile(.75)
Q1=popularity_d.quantile(.25)
IQR=Q3-Q1
lower_limit = Q1 -(1.5*IQR)
upper_limit = Q3 +(1.5*IQR)
popularity_outliers=popularity_d[(popularity_d<lower_limit)|(popularity_d>upper_limit)]
```

```python
popularity_outliers
```

```python
popularity_filtered=popularity_d[(popularity_d>lower_limit)&(popularity_d<upper_limit)]
```

```python
popularity_filtered
```

```python
vote_count_d=df['vote_count']
Q1=vote_count_d.quantile(.75)
Q3=vote_count_d.quantile(.25)
IQR=Q3-Q1
lower_limit=Q1-(IQR*1.5)
upper_limit=Q3+(IQR*1.5)
vote_count_outliers=vote_count_d[(vote_count_d<lower_limit)|(vote_count_d>upper_limit)]
```
